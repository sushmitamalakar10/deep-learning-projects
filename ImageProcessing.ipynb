**Converting RGB images to Grayscale image uisng OpenCV**

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('/content/cat.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```
